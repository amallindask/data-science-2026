```python
import pandas as pd
import numpy as np

# 1 Muat dasaset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
```

```python
# 2 Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum)
```

```python
# 3 filtering = Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') & df['Pclass']==1]
print(wk1['Survived'].mean(), wk1['Age'].mean())
```

```python
# 4 Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat= ('Survived','mean'),
).round(3)
print(hasil)
```

```python
# Q2 : KOLOM MANA YANG MEMILIKI NILAI KOSONG (MISSING)?
print(f"Missing value pada dataset: \n {df.isnull().sum()}")
print(f"\nMissing value dalam persentase(%): \n {df.isnull().sum()}")
```

```python
# Q3 : BERAPA PERSEN PENUMPANG YANG SELAMAT(SURVIVED=1)?
# selamat = wk1[wk1['Survived']==1]

print(f"Persentase penumpang wanita yang selamat di Kelas 1 : {wk1['Survived'].mean()*100} %")
```

```python
# Q4 : FILTERING PENUMPANG WANITA DARI KELAS 1 (Pclass=1). BERAPA JUMLAHNYA DAN BERAPA RATA RATA USIANYA?
# wk1 = df[(df['Sex']=='female') & df['Pclass']==1]
print(f"Jumlah penumpang wanita di Kelas 1 : {wk1.shape[0]} orang")
print(f"Rata-rata usia penumpang kelas 1 adalah : {wk1['Age'].mean()} tahun")
```

```python
# Q5 : GUNAKAN GROUPBY UNTUK MENGHITUNG TINGKAT KESELAMATAN (MEAN SURVUVED) PER KELAS PENUMPANG. KELAS MANA YANG PALING TINGGI?
print(f"Tingkat Keselamatan Per kelas :")
print(hasil)
```
