```python
import pandas as pd
import numpy as np

# 1 Muat dasaset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
df
```

```python
# 2 Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum)
```

```python
# 3 filtering = Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') & df['Pclass']==1]
print(wk1['Survived'].mean(), wk1['Age'].mean())
```

```python
# 4 Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat= ('Survived','mean'),
).round(3)
print(hasil)
```
